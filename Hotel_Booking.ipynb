# Analysis and Findings

```python
custom_palette = sns.color_palette('rocket')
plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(),edgecolor='k', width=0.7, color=custom_palette)
plt.show()
```

**This bar graph uncovers  a compelling story of reservations status distinguishing between cancellations and non-cancellations. While  67%  reservations stand firm, a significant 37% have sadly been cancelled, impacting our earnings. Let's uncover the reasons behind these cancellations and take proactive steps to secure our success.**



```python
custom_palette = sns.color_palette('rocket')

plt.figure(figsize=(8,5))
ax = sns.countplot(x='hotel',hue='is_canceled', data = df, palette = custom_palette ,edgecolor='k', width=0.7)
plt.title('Reservation status in different hotels')
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.show()
```

**In the realm of hotel bookings, City Hotels outshine Resort Hotels, potentially due to their AFFORDABILITY and ACCESSBILITY. By strategically addressing the pricing and emphasizing the unique offerings of Resort Hotels, we can captivate a wider audience and secure our position in unforgettable hospitality experiences.**

```python
resort_hotel=df[df['hotel']== 'Resort Hotel']
canceled_p = resort_hotel['is_canceled'].value_counts(normalize = True)
formatted_p = canceled_p.apply(lambda x: '{:.0%}'.format(x))
formatted_p
```

```python
city_hotel=df[df['hotel']== 'City Hotel']
canceled_c = city_hotel['is_canceled'].value_counts(normalize = True)
formatted_c = canceled_c.apply(lambda x: '{:.0%}'.format(x))
formatted_c
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(15, 7))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=20)
plt.plot(resort_hotel.index, resort_hotel['adr'], color='darkred', label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], color='yellow',label='City Hotel')
plt.legend(fontsize=15)

plt.show()
```

**The line graph above expose that  the average daily rate for city hotels occasionally falls below than that of resort hotels, with some days showcasing an even greater disparity. It's worth noting that weekends and holidays often lead to higher rates for resort hotels.**

```python
# Create a custom palette with dark pink shades
custom_palette = sns.color_palette('rocket')

plt.figure(figsize=(15, 7))
plt.title('Reservation Status Per Month', fontsize=20)

# Plot the countplot with the custom palette
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette=custom_palette, edgecolor='k', width=0.7)

plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['Not Canceled', 'Canceled'])

plt.show()
```

**The insightful grouped bar graph highlights the months with the highest and lowest reservation levels. Notably, August emerges as a month marked by both the highest number of confirmed reservations and cancellations, while January takes the spotlight as the month with the highest count of cancellations.**
